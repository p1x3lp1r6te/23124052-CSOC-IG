```python
import matplotlib.pyplot as plt
import numpy as np

# Define metrics for each method and split
methods = ['Pure Python', 'NumPy', 'Scikit-learn']

# MAE values
mae_train = [54364.47, 54364.47, 53517.63]
mae_val = [53864.67, 53864.67, 53034.09]
mae_test = [53742.84, 53742.84, 53225.16]

# RMSE values
rmse_train = [74210.12, 74210.12, 73620.87]
rmse_val = [72960.53, 72960.53, 72434.23]
rmse_test = [72499.63, 72499.63, 72006.72]

# R2 values
r2_train = [0.5902, 0.5902, 0.5967]
r2_val = [0.5966, 0.5966, 0.6024]
r2_test = [0.5990, 0.5990, 0.6044]

# Function to plot bar chart
def plot_metric(metric_values, title, ylabel):
    bar_width = 0.2
    x = np.arange(len(methods))

    plt.figure(figsize=(8, 5))
    plt.bar(x - bar_width, metric_values[0], width=bar_width, label='Train')
    plt.bar(x, metric_values[1], width=bar_width, label='Validation')
    plt.bar(x + bar_width, metric_values[2], width=bar_width, label='Test')

    plt.xticks(x, methods)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Plot all metrics
plot_metric([mae_train, mae_val, mae_test], 'MAE Comparison Across Methods', 'MAE')
```

```python
plot_metric([rmse_train, rmse_val, rmse_test], 'RMSE Comparison Across Methods', 'RMSE')
```

```python
plot_metric([r2_train, r2_val, r2_test], 'R² Score Comparison Across Methods', 'R²')
```
